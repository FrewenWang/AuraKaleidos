torch框架给我们干的最厉害的就是帮我们把反向传播计算完毕

```python
import torch

# 方法一
x = torch.randn(3, 4, requires_grad=True)
x
```

```python
x = torch.randn(3,4)
x.requires_grad = True
x
```

```python
b=torch.randn(3,4,requires_grad=True)
```

```python
t = x + b
```

```python
y = t.sum()
y
```

```python
y.backward()
```

```python
b.grad
```
