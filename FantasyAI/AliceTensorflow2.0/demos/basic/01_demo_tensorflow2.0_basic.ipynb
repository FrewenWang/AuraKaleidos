```python
# 张量可以是一个数组
value = tf.constant(1.0)
tf.print(value)
# 
# tf.constant(1.)

# tf.constant(1.0,dtype=tf.float64)
```

```python
# Numpy数组作为张量参数
tf.constant(np.array([1,2]))
tf.constant(np.array([1.0,2.0]))
tf.constant(np.array([1.0,2.0]),dtype=tf.float32)
```

```python
# tf.constant的函数，改变张量中元素的数据类型
# tf.cast(x,dtype)

a = tf.constant(np.array([1,2]))
b = tf.cast(a,dtype=tf.float32)
b.dtype

# 数据类型转换
# TODO 张量的数据和2.0有很大的不同
a = tf.constant(123456789,dtype=tf.int32)
tf.cast(a,tf.int16)

# 数据类型转换
# TODO 张量的数据和2.0有很大的不同
a = tf.constant([-1,0,1,2])
tf.cast(a,tf.bool)

# 数据类型转换
# 参数类型为字符串
tf.constant("Hello")
```

```python
# tf.convert_to_tensor(数组/列表/数字/布尔类型/字符串)
na = np.arange(12).reshape(3,4)
print(na)
ta = tf.convert_to_tensor(na)
type(na)
type(ta)

# is_tensor函数
tf.is_tensor(ta)

# isinstance函数
isinstance(ta,tf.Tensor)
isinstance(na,np.ndarray)
```
