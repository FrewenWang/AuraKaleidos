```python
# 我们来学习一下张量的运算

# 张量加减乘除运算、幂运算、对数运算、矩阵运算
# 基本数学运算
# tf.add(x,y)  #将x和y逐个元素相加
# tf.subtract(x,y)  #将x和y逐个元素相减
# tf.multiply(x,y)  #将x和y逐个元素相乘
# tf.divide(x,y)    #将x和y逐个元素相除
# tf.math.mod(x,y)    #将x和y逐个元素取模

# 减法、乘法、除法运算也都是这两个张量的逐个元素的运算，并且要求各个张量中的元素数据类型都一致。
a1 = tf.constant([0,1,2])
b1 = tf.constant([3,4,5])
tf.add(a1,b1 )
```

```python

a1 = tf.constant([0,1,2])
b1 = tf.constant([3,4,5])
tf.subtract(a1,b1 )
```

```python
a1 = tf.constant([0,1,2])
b1 = tf.constant([3,4,5])
tf.multiply(a1,b1 )
```

```python
a1 = tf.constant([0,1,2])
b1 = tf.constant([3,4,5])
tf.divide(a1,b1 )
```

```python
a1 = tf.constant([0,1,2])
b1 = tf.constant([3,4,5])
tf.math.mod(a1,b1 )
```

```python
# 基本数学运算

# 幂指数、指数、对数运算
# tf.pow(x,y)  #将x和y逐个元素相加
# tf.square(x,y)  #将x和y逐个元素相减
# tf.sqrt(x,y)  #将x和y逐个元素相乘
# tf.exp(x,y)    #将x和y逐个元素相除
# tf.math.log(x,y)    #将x和y逐个元素取模

x = tf.range(4)
x
tf.pow(x,2)

# 二维张量的运算
x1 = tf.constant([[2,2],[3,3]])
x2 = tf.constant([[8,16],[2,3]])
tf.pow(x1,x2)
```
