```python

# 鸢尾花数据集。是一个分类数据集
# 鸢尾花不是tensorflow和keras的数据

# keras.utils.get_file(fname,origin,cachedir)
# fname:下载后的文件名
# origin:文件的URL地址
# cache_dir:下载后文件的存储位置
# 返回值：是下载后的文件在本地磁盘的绝对路径

# 下载鸢尾花的训练数据集
TRAIN_URL = "http://download.tensorflow.org/data/iris_training.csv"
fname = TRAIN_URL.split('/')[-1]
print("下载后的名称:",fname)
train_path = keras.utils.get_file(fname,TRAIN_URL)
print("下载后的路径:",train_path)
```

```python
# Pandas二维数据基本操作
# 读取CSV数据集文件
# pd.read_csv(filepath_or_buffer,header,names)

# 
# header 可以执行数据表中的
pd.read_csv(train_path)
dataframe_iris = pd.read_csv(train_path)

print(type(dataframe_iris))
```

```python
# 将Pandas二维数组转化为Numpy数组
iris_np = np.array(dataframe_iris)
# 下面两种方法都可以将dataframe_iris转化为numpy数组
iris_np = dataframe_iris.values
# iris_np = dataframe_iris.as_matrix()

print(type(dataframe_iris))

print(type(iris_np))


print(iris_np.shape)
```

```python
# 通过numpy的数组来操作鸢尾花数据
# 读取前六行数据
iris_np[0:6]

iris_np[0:6,0:4]


# 得到所有数据中鸢尾花的种类
iris_np[:,4]
```
