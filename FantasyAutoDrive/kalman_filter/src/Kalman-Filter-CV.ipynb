## 车辆状态向量

自车运动的匀速模型

$$ x_k= \left[ \matrix{ x \\ y \\ \dot x \\ \dot y} \right] = \matrix{ \text{车辆的X轴坐标 X} \\ \text{车辆Y轴坐标 Y} \\
\text{车辆X方向的速度分量 in X} \\ \text{车辆Y方向的速度分享 in Y}}$$

车辆的动力学模型：

$$x_{k+1} = \textbf{A} \cdot x_{k}$$

带入上面的方程

$$x_{k+1} = \begin{bmatrix}1 & 0 & \Delta t & 0 \\ 0 & 1 & 0 & \Delta t \\ 0 & 0 & 1 & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix} \cdot \begin{bmatrix} x \\ y \\ \dot x \\ \dot y \end{bmatrix}_{k}$$



观测模型

$$y = \textbf{H} \cdot x$$
