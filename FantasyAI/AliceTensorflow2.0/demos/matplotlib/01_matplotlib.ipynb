```python
# Matplotlib绘图库
# Figure对象。相当于一张画布，下面我们学习怎么实例化Figure对象
# figure(num,figsize,dpi,facecolor,edgecolor,frameon)
# 记住所有的参数都是可选的，我们学习一下参数说明：
# num: 图像编号或者名称，取值为数组/字符串
# figsize：绘图对象的宽和高，单位是英寸
# dpi：绘图对象的分辨率，缺省值为80
# facecolor：背景颜色
# edgecolor：边框颜色
# frameon：表示是否显示边框

# plt.plot()  #绘制一个空白图形

# plt.show()  #让图形显示出来

# 下面我们演示一个Demo：
plt.figure(figsize=(3,2),facecolor="green",edgecolor="red",frameon=True)
plt.plot()
plt.show()
```

```python
# 鸢尾花数据集。是一个分类数据集
# 鸢尾花不是tensorflow和keras的数据

# keras.utils.get_file(fname,origin,cachedir)
# fname:下载后的文件名
# origin:文件的URL地址
# cache_dir:下载后文件的存储位置
# 返回值：是下载后的文件在本地磁盘的绝对路径

# 下载鸢尾花的训练数据集
TRAIN_URL = "http://download.tensorflow.org/data/iris_training.csv"
fname = TRAIN_URL.split('/')[-1]
print("下载后的名称:",fname)
train_path = keras.utils.get_file(fname,TRAIN_URL)
print("下载后的路径:",train_path)

```

```python

# header 可以执行数据表中的
pd.read_csv(train_path)
dataframe_iris = pd.read_csv(train_path)

ny_iris = np.array(dataframe_iris)

ny_iris[:,2]

ny_iris[:,3]
```

```python
# 我们开始实现画图的实现鸢尾花数据可视化

plt.figure(figsize=(12,12))
# plt.scatter(ny_iris[:,2],ny_iris[:,3])

# 我们需要进行不同品种的鸢尾花使用不同颜色
# 进行色彩映射
# x = np.arange(10)
# y = np.arange(10)
# dot_color=[0,1,2,0,1,2,2,1,1,0]
# plt.scatter(x,y,c=dot_color,cmap='brg')
# plt.show()

# 下面我们来看一下鸢尾花的数据显示
plt.figure(figsize=(12,12))
plt.scatter(ny_iris[:,2],ny_iris[:,3],c=ny_iris[:,4],cmap='brg')
plt.title("Anderson's Iris Data Set \n Blue->Seta=osa")
```
